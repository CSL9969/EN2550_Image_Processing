```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread(r'images/spider.png')
assert img is not None
fig , ax = plt.subplots(1,5)
fig.set_figwidth(20)
ax[0].imshow(cv.cvtColor(img , cv.COLOR_BGR2RGB))
ax[0].axis('off')
ax[0].title.set_text('original')


gamma = [0.2 , 0.8 , 1.2 , 2]

for i , g in enumerate(gamma):
    t = np.array([(p/255) ** g * 255 for p in range(256)]).astype(np.uint8)
    modified = cv.LUT(img, t)
    ax[i+1].imshow(cv.cvtColor(modified , cv.COLOR_BGR2RGB))
    ax[i+1].axis('off')
    ax[i+1].title.set_text('$\gamma$ = '+ str(g))

plt.show()



```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread(r'images/spider.png')
assert img is not None

t1 = np.linspace(0,100,50)
t2 = np.linspace(100,255,150)
t3 = [255]*56

t = np.concatenate((t1,t2,t3), axis = 0).astype(np.uint8)
fig , ax = plt.subplots(1,2)
fig.set_figwidth(15)
ax[0].plot(t)
modified = cv.LUT(img , t)
ax[1].imshow(cv.cvtColor(modified , cv.COLOR_BGR2RGB))
ax[1].axis('off')

```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread(r'images/shells.tif' , cv.IMREAD_GRAYSCALE)
assert img is not None

histogram = cv.calcHist([img] , [0], None,[256], [0,256])
fig , ax = plt.subplots(1,3)
fig.set_figwidth(20)
ax[0].plot(histogram)
ax[0].title.set_text('original histogram')

equalized = cv.equalizeHist(img)
equ_hist = cv.calcHist([equalized] , [0], None, [256], [0,256])
ax[1].plot(equ_hist)
ax[1].title.set_text('equalized histogram')

ax[2].imshow(equalized , cmap = 'gray' , vmin = 0, vmax = 255)
ax[2].axis('off')
plt.show()
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread(r'images/zion_pass.jpg')
assert img is not None

fig , ax = plt.subplots(1 , 2)
fig.set_figwidth(20)
fig.set_figheight(15)
ax[0].imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
ax[0].axis('off')

hsv_img = cv.cvtColor(img, cv.COLOR_BGR2HSV)
(h , s , v) = cv.split(hsv_img)
value = 50
s = cv.add(s , value)
hsv_img = cv.merge([h, s, v])
ax[1].imshow(cv.cvtColor(hsv_img, cv.COLOR_HSV2RGB))
ax[1].axis('off')

plt.show()


```
